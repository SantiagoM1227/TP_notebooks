Nous allons produire un réseau de neurones autoencodeur sur la base de données MNIST. L'objectif d'un autoencodeur est d'être capable, à l'aide de la première partie du réseau appelée encodeur, de réduire la dimension des données de manière non linéaire dans un espace appelé espace latent.
